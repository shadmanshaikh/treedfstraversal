```python
def convert_tree_values_to_integers(node):
    if node is None:
        return None
    
    # Convert the current node's value to an integer
    node.value = int(node.value)
    
    # Recursively convert values in left and right subtrees
    convert_tree_values_to_integers(node.left)
    convert_tree_values_to_integers(node.right)
```
